```python
import gc
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pd.set_option('display.max_columns', 1000)
```

```python
order_9 = pd.read_csv("../data/order_info_201909_WHsec.csv")  # 171657
order_9 = order_9.iloc[:, 1:]
order_10 = pd.read_csv("../data/order_info_201910_WHsec.csv")  # 207130
order_10 = order_10.iloc[:, 1:]
order_11 = pd.read_csv("../data/order_info_201911_WHsec.csv")  # 295485
order_11 = order_11.iloc[:, 1:]
```

```python
order = pd.concat([order_9, order_10, order_11], axis=0)  # 674272

del order_9, order_10, order_11
gc.collect()
```

```python
order.rename(
    columns={
        'renter_id': 'passenger_id', 
        'id_': 'order_id', 
        'receive_lng': 'receive_longitude', 
        'receive_lat': 'receive_latitude', 
        'driver_arrive_appoint_lng': 'driver_arrive_appoint_longitude', 
        'driver_arrive_appoint_lat': 'driver_arrive_appoint_latitude',
        'go_on_lng': 'get_on_longitude', 
        'go_on_lat': 'get_on_latitude', 
        'close_lng': 'close_longitude', 
        'close_lat': 'close_latitude', 
    }, 
    inplace=True
)
```

```python
order['longitude'] = np.round(order.get_on_longitude, decimals=3)
order['latitude'] = np.round(order.get_on_latitude, decimals=3)
```

```python
order_counts = order.groupby(['longitude', 'latitude'])[['order_id']].count()
order_counts = order_counts.reset_index().rename(columns={'order_id': 'demand'})
```

```python
order_counts
```

## 2 画热力图

```python
import folium
import webbrowser
from folium.plugins import HeatMap
```

```python
latitudes = list(order_counts.latitude)
longitudes = list(order_counts.longitude)
weights = list(order_counts.demand)
llws = [[latitudes[i], longitudes[i], weights[i]] for i in range(len(longitudes))]
```

```python
map_osm = folium.Map(location=[35, 110], zoom_start=5)
HeatMap(llws).add_to(map_osm)
```

```python
html_filepath = "../result/location-demand.html"
map_osm.save(html_filepath)
```
